Загрузим датасет о ценах на дома в Бостоне. 

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.data.shape)
```

Каждый дом - это вектор в 13-мерном пространстве. 
Написать код для подсчета скалярного произведения двух первых домов из датасета.

```python
x1 = boston.data[0] # дом 1
x2 = boston.data[1] # дом 2
```

```python
import numpy as np
np.dot(x1,x2)
```

этап 2: взять два атрибута и посчитать скалярное произведение двух векторов (2 дома), начертить 2 дома как вектора в двухмерном пространстве, и сравниваем графически и через скалярное произведение.


```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(x1[0], x1[1], 'o')
plt.plot(x2[0], x2[1], 'o')
plt.grid()
plt.show()
```

**По графику видно, что два дома по двум критериям кардинально отличаются друг от друга**

```python
np.dot(x1[:2], x2[:2])
```

**Скалярное произведение практически равно 0, что говорит нам о том, что вектора перпендикулярны.** 

Нарисовать два дома в 3-мерном пространстве (атрибуты выбрать самостоятельно)

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D

fig = pylab.figure()
ax = Axes3D(fig)
ax.scatter(x1[0], x1[1], x1[2])
ax.scatter(x2[0], x2[1], x2[2])

pylab.show()
```

Найти косинус угла между векторами x1 и x2, где x1 и x2 - это два дома из 2ух атрибутов, полученных на этапе 2.

```python
import math

def dlina_vectora(v):
    result = 0
    for x in v:
        result += x ** 2
    
    return math.sqrt(result)    

cos = np.dot(x1[:2], x2[:2]) / (dlina_vectora(x1[:2]) * dlina_vectora(x2[:2]))
print('Косинус угла ', cos)
```

Найти расстояние между точками x1 и x2.

```python
math.sqrt((x2[0] - x1[0])**2 + (x2[1] - x1[1])**2)
```

Задан объект в пространстве 2D. 
Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).

Вращаем обратно

```python
new_A = np.column_stack((x,y))
u = math.radians(-45)
um = [[math.cos(u), -math.sin(u)], [math.sin(u), math.cos(u)]]

new_A = np.dot(new_A,um)

x = new_A[:,0]
y = new_A[:,1]

plt.plot(x, y)

plt.plot([-200,200], [-200,200])

plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1,2)

```python
# смещаем полученную фигуру
new_A = np.column_stack((x,y))

delta = [200, 300]

new_A = new_A + delta

size = [0.5, 1.2]

new_A = new_A * size

x = new_A[:,0]
y = new_A[:,1]

plt.plot(x, y)
```
